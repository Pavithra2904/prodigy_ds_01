```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
data = pd.read_csv("/content/sample_data/population.csv")

# Plotting the distribution of average household rate by location (categorical variable)
plt.figure(figsize=(10, 6))
sns.barplot(x="PROVINCE", y="AVG HOUSEHOLD SIZE (RURAL)", data=data)
plt.title("Distribution of Average Household Rate by Location")
plt.xlabel("PROVINCE")
plt.ylabel("AVG HOUSEHOLD SIZE (RURAL)")
plt.show()

# Plotting the histogram of annual growth rate (continuous variable)
plt.figure(figsize=(10, 6))
sns.histplot(data["ANNUAL GROWTH RATE (RURAL)"], bins=20, kde=True)  # Adjust bins as needed
plt.title("Histogram of Annual Growth Rate")
plt.xlabel("ANNUAL GROWTH RATE (RURAL)")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("/content/sample_data/population.csv")

# Filter data for urban and rural areas
urban_data = data[data["PROVINCE"] == "Urban"]
rural_data = data[data["PROVINCE"] == "Rural"]

# Calculate average annual growth rate for urban and rural areas within each province
urban_avg_growth_rate = urban_data.groupby("PROVINCE")["ANNUAL GROWTH RATE (URBAN)"].mean()
rural_avg_growth_rate = rural_data.groupby("PROVINCE")["ANNUAL GROWTH RATE (RURAL)"].mean()

# Create a bar chart
provinces = urban_avg_growth_rate.index
bar_width = 0.35
index = range(len(provinces))

plt.figure(figsize=(10, 6))
plt.bar(index, urban_avg_growth_rate, bar_width, label='Urban', color='blue')
plt.bar([i + bar_width for i in index], rural_avg_growth_rate, bar_width, label='Rural', color='green')

plt.xlabel('PROVINCE')
plt.ylabel('Average Annual Growth Rate')
plt.title('Average Annual Growth Rate by Province and Location Type')
plt.xticks([i + bar_width / 2 for i in index], provinces)
plt.legend()
plt.tight_layout()
plt.show()
```
